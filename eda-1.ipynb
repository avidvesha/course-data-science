```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
null_data = df[df.isnull().any(axis=1)]
null_data
```

```python
df.loc[10, 'Harga Makanan'] = 14000
df.loc[15, 'Harga Minuman'] = 3000
df['Review'].fillna(df['Review'].median(), inplace=True)
df
```

```python
print(df.iloc[10])
print(df.iloc[15])
print(df.iloc[20])
print(df.iloc[23])
```

```python
df.duplicated().sum()
```

```python
df_duplicates = df[df.duplicated(subset='ID Pelanggan')]
df_duplicates
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.info()
```

```python
df['Review'].value_counts()
```

```python
meanMedianHarga = df[['Harga Makanan', 'Harga Minuman']].agg([np.mean, np.median])
meanMedianHarga
```

```python
range = df['Total Bayar'].max() - df['Total Bayar'].min()
print(f"Total bayar tertinggi : {df['Total Bayar'].max()}")
print(f"Total bayar terendah : {df['Total Bayar'].min()}")
print(f"Range pembayaran : {range}")
```

```python
boxPlotData = df['Total Bayar']
plt.title("Price Distribution")
plt.boxplot(boxPlotData, patch_artist=True, labels=['Total Bayar'])
plt.show()
```

```python
df['Makanan'].value_counts()
```

```python
replaceFood = {
  'Ayam bakar': 0,
  'Ayam geprek': 1,
  'Ayam goreng': 2,
  'Lele bakar': 3,
  'Lele goreng': 4,
  'Nila bakar': 5,
  'Nila goreng': 6,
}
df = df.replace({'Makanan': replaceFood})
df.head()
```

```python
df['Minuman'].value_counts()
```

```python
replaceDrink = {
  'Es buah': 0,
  'Es teh': 1,
  'Es teler': 2,
  'Es jeruk': 3,
  'Jus': 4,
}
df = df.replace({'Minuman': replaceDrink})
df.head()
```

```python
df.corr(numeric_only=True).style.background_gradient(cmap='gist_gray')
```

```python
matrixCorr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(matrixCorr, annot=True, cmap='coolwarm')
plt.title("Correlations Matrix")
plt.show()
```
